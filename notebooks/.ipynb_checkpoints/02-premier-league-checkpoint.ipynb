### **Goals**

```python
plt.xticks(rotation=90);
sns.barplot(x="season",y="goals",data=df_stats)
plt.show()

plt.xticks(rotation=90);
sns.barplot(x="team",y="goals",data=df_stats)
plt.show()
```

### **Penalty Save**

```python
tm=df_stats.groupby(['team'])
tm=tm['penalty_save'].agg('max')
tm=pd.DataFrame(tm)
tm.reset_index(drop=False,inplace=True)
tm.set_index('penalty_save',inplace=True)
tm.sort_index(ascending=False)
tm.reset_index(drop=False,inplace=True)
tm=tm.nlargest(15,"penalty_save")
```

```python
plt.xticks(rotation=90);
sns.scatterplot(x="team",y="penalty_save",data=tm)
plt.show()
```

### **Touches for Team**

```python
tm=df_stats.groupby(['team'])
tm=tm['touches'].agg('max')
tm=pd.DataFrame(tm)
tm.reset_index(drop=False,inplace=True)
tm.set_index('touches',inplace=True)
tm.sort_index(ascending=False)
tm.reset_index(drop=False,inplace=True)
tm=tm.nlargest(15,"touches")
```

```python
plt.xticks(rotation=90);
sns.scatterplot(x="team",y="touches",data=tm)
plt.show()
```

### **Yellow Card**

```python
yel=df_stats.groupby('team').total_yel_card.mean()
yel=pd.DataFrame(yel)
yel.sort_values('total_yel_card',ascending=False,inplace=True)
```

```python
sns.heatmap(yel,cmap="Wistia_r")
plt.show()
```

## **Red Card**

```python
red=df_stats.groupby('team').total_red_card.mean()
red=pd.DataFrame(red)
red.sort_values('total_red_card',inplace=True)
```

```python
sns.heatmap(red,cmap="Reds")
plt.show()
```
